## Problem 4. Wave dispersion

Make the derivative and the expansion to see the leading order

```python
def eta(x,x0,lam):
    A = 0.1
    dif = x-x0
    fun = np.zeros(len(x))
    for i in range(len(x)):
        if -0.5*lam < dif[i] < 0.5*lam:
            fun[i] = 2*A*np.cos(np.pi*dif[i]/lam)**2
    return fun
```
